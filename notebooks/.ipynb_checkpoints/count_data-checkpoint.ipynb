```python
import pandas as pd
from PIL import Image
import numpy as np
import io
import os

import tensorflow as tf
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras import Model
from pyspark.sql.functions import col, pandas_udf, PandasUDFType, element_at, split, udf
from pyspark.sql import SparkSession
from pyspark.ml.linalg import Vectors, VectorUDT
from pyspark.sql.types import ArrayType, FloatType
from pyspark.ml.feature import PCA, StandardScaler
```

```python
print(os.getcwd())
```

```python
PATH = os.getcwd()
PATH_Data = PATH+'/fruits-360_dataset/fruits-360/Test'
print('PATH:            '+PATH+\
      '\nPATH_Data:       '+PATH_Data)
```

```python
spark = (SparkSession
             .builder
             .appName('P8')
             .master('local')
             .config("spark.sql.parquet.writeLegacyFormat", 'true')
             .getOrCreate()
)
```

```python
sc = spark.sparkContext
```

```python
images = spark.read.format("binaryFile") \
  .option("pathGlobFilter", "*.jpg") \
  .option("recursiveFileLookup", "true") \
  .load(PATH_Data)
```

```python
images = images.withColumn('label', element_at(split(images['path'], '/'),-2))
print(images.printSchema())
print(images.select('path','label').show(5,False))
```

```python
from pyspark.sql.functions import countDistinct
images.select(countDistinct('label')).show()
```

```python
PATH_Data_test = PATH+'/fruits-360_dataset/fruits-360/Test_2'
```

```python
images_2 = spark.read.format("binaryFile") \
  .option("pathGlobFilter", "*.jpg") \
  .option("recursiveFileLookup", "true") \
  .load(PATH_Data_test)
```

```python
images_2 = images_2.withColumn('label', element_at(split(images_2['path'], '/'),-2))
print(images_2.printSchema())
print(images_2.select('path','label').show(5,False))
```

```python
images_2.groupby('label').count().show()
```

```python
import random
from PIL import Image
import matplotlib.pyplot as plt

# Récupérer 3 chemins d'images aléatoires
paths = images_2.select('path').rdd.map(lambda row: row['path']).collect()
sampled_paths = random.sample(paths, 3)
cleaned_paths = [p.replace("file:", "") for p in sampled_paths]

# Affichage
plt.figure(figsize=(12, 4))
for i, p in enumerate(cleaned_paths):
    img = Image.open(p)
    plt.subplot(1, 3, i + 1)
    plt.imshow(img)
    plt.axis('off')
plt.show()
```
